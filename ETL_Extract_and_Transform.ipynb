```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "CSVFILES/Asthma_by_zip.csv"
asthma_visit_by_zip_df = pd.read_csv(csv_file)
asthma_visit_by_zip_df.head()
```

```python
zip_codes = asthma_visit_by_zip_df["ZIP code"].str.slice(0,5)
asthma_visit_by_zip_df["ZIP code"]= zip_codes
```

```python
asthma_visit_by_zip_df
```

```python
asthma_grouped = asthma_visit_by_zip_df.groupby("County")
aaa = round(asthma_grouped.mean(),0)
aa2 = aaa.reset_index()
asthma_updated_2 = aa2[["County","Number of Visits"]]
asthma_updated_3 = asthma_updated_2.rename(columns = {"Number of Visits": "ER_Visits"})
asthma_updated_3
```

```python
asthma_title = asthma_visit_by_zip_df["County"].str.title()
asthma_visit_by_zip_df["County"] =asthma_title 

asthma_updated = asthma_visit_by_zip_df.rename(columns={"Age Group":"Age_Group", "Number of Visits": "Number_of_Visits","ZIP code":"Zip_Code"})



```

```python
# data set # 2
csv_file = "CSVFILES/Air_Quality_byZip.csv"
Air_Quality_by_Zip_df = pd.read_csv(csv_file)
```

```python
AQS_by_county= Air_Quality_by_Zip_df[["Date","DAILY_AQI_VALUE","COUNTY"]]
AQS_by_county.head(15)
```

```python
mean_by_AQS = round(AQS_by_county.groupby('COUNTY').mean(),0)
mean_by_AQS_1=mean_by_AQS.reset_index()
mean_by_AQS_1
```

```python
rds_connection_string = "username:password@localhost:5432/ETL"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
asthma_updated_3.to_sql(name='asthma_updated', con=engine, if_exists='append', index=False)
```

```python
mean_by_AQS_1.to_sql(name='aqi_by_mean', con=engine, if_exists='append', index=False)
```
